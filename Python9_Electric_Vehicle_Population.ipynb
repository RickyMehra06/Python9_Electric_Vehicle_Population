### Python 9: Electric Vehicle Population Data

Write a program to read the data from the following link, perform data analysis and answer the following questions:

Link - https://data.wa.gov/api/views/f6w7-q2d2/rows.csv?accessType=DOWNLOAD

```python
import pandas as pd
df = pd.read_csv("https://data.wa.gov/api/views/f6w7-q2d2/rows.csv?accessType=DOWNLOAD")
df.head()
```

```python
df.columns
```

#### Get all the cars and their types that do not qualify for clean alternative fuel vehicle

```python
df["Clean Alternative Fuel Vehicle (CAFV) Eligibility"].value_counts()
```

```python
df[df["Clean Alternative Fuel Vehicle (CAFV) Eligibility"] != 
                                                   "Clean Alternative Fuel Vehicle Eligible"][["Make","Electric Vehicle Type" ]]
```

#### Get all TESLA cars with the model year, and model type made in Bothell City

```python
# Get all TESLA cars with the model year, and model type made in Bothell City.

df[(df["Make"]=="TESLA") & (df["City"] == "Bothell")][['Model Year', 'Model']]
```

#### Get all the cars that have an electric range of more than 100, and were made after 2015

```python
# Get all the cars that have an electric range of more than 100, and were made after 2015

df[(df["Electric Range"] > 100) & (df["Model Year"] > 2015)]
```

#### Draw plots to show the distribution between city and electric vehicle type

```python
# Draw plots to show the distribution between city and electric vehicle type

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(7,3))
sns.countplot(data=df, x='Electric Vehicle Type')
plt.ylabel('Count')
plt.title('Distribution of Electric Vehicle Types by City')
plt.show()
```
